```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv" 
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
# District Summary
# Calculate the total number of schools
# Calculate the total number of students
# Calculate the total budget
# Calculate the average math score
# Calculate the average reading score
# Calculate the percentage of students with a passing math score (70 or greater)
# Calculate the percentage of students with a passing reading score (70 or greater)
# Calculate the percentage of students who passed math and reading (% Overall Passing)
# Create a dataframe to hold the above results
# Optional: give the displayed data cleaner formatting

# Calculate the number of schools
total_schools = schldat_description_df.loc["count","size" ]

# Calculate the total number of students
total_students = studat_description_df.loc["count","Student ID" ]

# * Average math score
average_math = student_data["math_score"].mean()

# * Average reading score
average_reading = student_data["reading_score"].mean()

# * % passing math (the percentage of students who passed math)
math_stupassed = student_data.loc[student_data["math_score"]>=70,:]
math_passed = len(math_stupassed)/total_students*100

# * % passing reading (the percentage of students who passed reading)
#reading_passed = student_data
read_stupassed = student_data.loc[student_data["reading_score"]>=70,:]
read_passed = len(read_stupassed)/total_students*100

# * % overall passing (the percentage of students who passed math AND reading)
#overall_pass

#multiple criteria
passed_math_and_reading = student_data.loc[(student_data["math_score"] > 69) & (student_data["reading_score"] > 69), :]
passed_math_and_reading = len(passed_math_and_reading) / total_students * 100



# ______________________________District Summary View___________________________

# Consolicate for DISTRICT SUMMARY dataframe: 


district_summary_df = pd.DataFrame({"Total Schools": [total_schools],
                               "Total Students": [total_students],
                               "Total Budget": [total_budget],
                               "Average math score": [average_math],
                               "Average reading score": [average_reading],
                               "% Passing Math": [math_passed],
                               "% Passing Reading": [read_passed],
                               "% Passing Overall": [passed_math_and_reading]})

# district_summary_df = district_summary_df[[
#     "Total Schools",
#     "Total Students",
#     "Total Budget",
#     "Average Math Score",
#     "Average Reading score",
#     "% Passing Math",
#     "% Passing Reading",
#     "% Passing Overall"
# ]]

district_summary_df["Total Students"] = district_summary_df["Total Students"].map("{:,}".format)
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("${:,.2f}".format)

district_summary_df

```

```python
### School Summary

# Create a DataFrame that summarizes key metrics about each school, including the following:

# * School name
# * School type
# * Total students
# * Total school budget 
# * Per student budget
# * Average math score
# * Average reading score
# * % passing math (the percentage of students who passed math)
# * % passing reading (the percentage of students who passed reading)
# * % overall passing (the percentage of students who passed math AND reading)


# Determine the School Type
school_types = school_data.set_index(["school_name"])["type"]


#Total Students
school_total_students = school_data.set_index(["school_name"]) ["size"]

#Total School Budget
total_school_budget = school_data.set_index(["school_name"]) ["budget"]

#per student budget by school
per_school_budget = school_data_complete.groupby(["school_name"]).mean()["budget"]
per_student_budget = per_school_budget / school_total_students

# Average math per school
school_average_math = school_data_complete.groupby(["school_name"]).mean()["math_score"]


# Average reading per school
school_average_reading = school_data_complete.groupby(["school_name"]).mean()["reading_score"]


# Percentage of those who passed reading, math, and overall (both)

#     Start by getting the count of students who passed r, m, o
#     Students who passed math
school_passing_math = school_data_complete[(school_data_complete["math_score"] >= 70)]
#     Students who passed reading
school_passing_reading = school_data_complete[(school_data_complete["reading_score"] >= 70)]
#     Students who passed overall
school_passing_overall = school_data_complete[(school_data_complete["reading_score"] >= 70)
                                                & (school_data_complete["math_score"] >= 70)]


#     Percent calculations
pct_school_passing_math = school_passing_math.groupby(["school_name"]).count()["student_name"] / school_total_students * 100
pct_school_passing_reading = school_passing_reading.groupby(["school_name"]).count()["student_name"] / school_total_students * 100
pct_school_passing_overall = school_passing_overall.groupby(["school_name"]).count()["student_name"] / school_total_students * 100


# ______________________________School Summary View___________________________

# SCHOOL SUMMARY: 

# Consolidate to dataFrame
per_school_summary = pd.DataFrame({"School Type": school_types,
                                   "Total Students": school_total_students,
                                   "Total School Budget": total_school_budget,
                                   "Per Student Budget": per_student_budget,
                                   "Average Math Score": school_average_math,
                                   "Average Reading Score": school_average_reading,
                                   "% Passing Math": pct_school_passing_math,
                                   "% Passing Reading": pct_school_passing_reading,
                                   "% Overall Passing": pct_school_passing_overall})

# Minor data wrangling
per_school_summary = per_school_summary[[
    "School Type", "Total Students", "Total School Budget", "Per Student Budget",
    "Average Math Score", "Average Reading Score", 
    "% Passing Math", "% Passing Reading", 
    "% Overall Passing"
]]


per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.2f}".format)
per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)

# Display the DataFrame
per_school_summary



```
